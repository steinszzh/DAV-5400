## Part 2: Data cleaning

```python
#import extra package
import csv
```

```python

#makng a list to reorganize and clean data

with open('tournamentinfo.txt', 'r') as f:
  reader = csv.reader(f)
  my_list = list(reader)

print(my_list)
```

```python
#Test couple of data points
my_list[4]
my_list[7]
#check the length of whole list
length=len(my_list)
length
#Got 196 list and it matched earlier 196 lines.
#check whether it is last player on line 193. And it is verified.
my_list[193]
```

By analyzing the pattern, we can tell that starting from fourth line, every three list is a new player 's total points and round data.Also, next line(list) is corresponding location,ID and rating data. In addition, previous read text data is verified here. So a general algorithm can be generated. 

```python
#Firstly, create player with total points, and round only. 
#Based on analysis, we can tell that the distance between two player data is two 

#using np arange to make spaced value. In this case, since we know the stop number of the line is 196.
row_num = np.arange(1,196,3)
step = 0
#create an empty player only list
player_only = []

#using for loop to get player only data
for i in my_list:
    if(step in row_num):
        player_only.append(i)
    step = step+1
#get player only data
player_only 
```

After getting player only data, we can use similar algorithm to get location and rating data.

```python

#Similar algorithm and revised for this 

#using np arange to make spaced value. In this case, since we know the stop number of the line is 196.
row_num = np.arange(2,196,3)
step = 0
#create an empty loc_rating list
loc_rating = []

#using for loop to get loc_rating data
for i in my_list:
    if(step in row_num):
        loc_rating.append(i)
    step = step+1
#get loc_rating data
loc_rating

#Since there is no pair number ,using length funcion to check whether there is missing value or not.
```

After getting these two list values, regular expression can be used to find names,rating,state and points. 
Firstly, I start with the name of regular expression

```python
# create a regex for name
name_re = re.compile("[A-Za-z]+\s[A-Za-z]+")
# use findall function to find all names
name = name_re.findall(str(player_only))
#print name
name
```

It is noticeable Player Name is also in this name list. One more step need to take to remove it. 

```python
#remove strings from name list 
name.remove('Player Name')
name
```

Then we can create the total points data.

```python
#Using similar method
# create a regex for total points
totalpt_re = re.compile("[0-9]\.[0-9]")
# use findall function to find all total points
totalpt= totalpt_re.findall(str(player_only))
#print totalpt
totalpt
```

Then using similar method to create the state data.

```python
#Using similar method
# create a regex for state
state_re = re.compile("[A-Z]{2}")
# use findall function to find all state
state= state_re.findall(str(loc_rating))
#print state
state
```

Also, there is three more wrong strings(US,CF and ID) was found,since first state was supposed to be ON.

```python
#remove three strings from list. 
#duplicate_state=('US','CF','ID') hard way of removing
state.remove('US')
state.remove('CF')
state.remove('ID')
state
```

Then using similar method to create the pre tournament rating data.

```python
#Using similar method
# create a regex for pre tournament rating add : to avoid also get post tournament value
prerating_re = re.compile("\:\s+[0-9]{3,4}")
# use findall function to find all state
prerating= prerating_re.findall(str(loc_rating))
#print state
prerating
```

Since it still has special character, another step need to take

```python
#clean the data
cleanup_re = re.compile("[0-9]{3,4}")
# use findall function to find all state
pre_rating= cleanup_re.findall(str(prerating))
#print state
pre_rating
```

## Create Dataframe

```python

table= {'Player Name':name,'Total Point':totalpt, 'State':state,'Pre_T_Rating':pre_rating} 
# Create DataFrame 
dataframe = pd.DataFrame(table) 
dataframe
```
